```python
"""
AUFGABE: 

Wann ist in Zukunft mit einem Pferd zu rechnen, welches schneller / gleich schnell wie Secretariat ist?
-> Exponentialverteilung
"""

# Get fastest time
fastest = belmont.loc[belmont["Miliseconds"].idxmin()]

# Occurences of fastest time
occurences = sum(1 for time in belmont["Miliseconds"] if time == fastest["Miliseconds"])

# Get the time delta of data set
belmont["Year"] = pd.to_datetime(belmont["Year"], format="%Y")
ly = belmont["Year"].min()
hy = belmont["Year"].max()
delta = int((hy - ly).days / 365)
print(F"Lowest year: {ly}  Highest year: {hy}  Delta: {delta}")

# Lambda = occurences / year
tau = occurences / delta

# Get exponential 
faster_times = np.random.exponential(tau, 100000)

# Plot the PDF and label axes
#plt.figure(figsize=(12,6))
_ = plt.hist(faster_times,
             bins=50, density=True, histtype='step')
_ = plt.xlabel('Years between faster times')
_ = plt.ylabel('PDF')

# Show the plot
plt.show()
```
